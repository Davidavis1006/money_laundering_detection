# ccba

```python
ccba = pd.read_csv('traindata/public_train_x_ccba_full_hashed.csv')
print(ccba.isnull().sum())
ccba
```

# clamt本月分期預借現金金額, csamt本月預借現金金額, cucah本月借現金額

```python
#平均相同ID的所有data
newccba = ccba.groupby(['cust_id']).agg({'lupay': np.mean,'byymm': np.mean,'cycam': np.mean,'usgam': np.mean,'clamt': np.mean,
                                         'csamt': np.mean,'inamt': np.mean,'cucsm': np.mean,'cucah': np.mean,})
newccba = newccba.drop ( 'byymm' , axis=1 )
newccba = newccba.astype(int)
newccba
```

```python
newccba['clamt'].value_counts().sort_index()
#newccba['csamt'].value_counts().sort_index()
#newccba['cucah'].value_counts().sort_index()
```

```python
#非0值改成1
clamt_mapping = { 0 : 0,"else":1}
newccba['clamt'] = newccba['clamt'].map(clamt_mapping).fillna(1).astype(int)
csamt_mapping = { 0 : 0,"else":1}
newccba['csamt'] = newccba['csamt'].map(csamt_mapping).fillna(1).astype(int)
cucah_mapping = { 0 : 0,"else":1}
newccba['cucah'] = newccba['cucah'].map(cucah_mapping).fillna(1).astype(int)

newccba['clamt'].value_counts()
newccba['csamt'].value_counts()
#newccba['cucah'].value_counts()
```

```python
newccba['lupay'].max()
newccba.loc[ newccba['lupay'] <= 100000, 'lupay' ] = 0 
newccba.loc[ (newccba['lupay'] > 100000) & (newccba['lupay'] <= 1000000), 'lupay' ] = 1 
newccba.loc[ (newccba['lupay'] > 1000000) & (newccba['lupay'] <= 5000000), 'lupay' ] = 2 
newccba.loc[ newccba['lupay'] > 5000000, 'lupay' ] = 3
newccba['lupay'].value_counts()
```

```python
newccba['usgam'].max()
newccba.loc[ newccba['usgam'] <= 100000, 'usgam' ] = 0 
newccba.loc[ (newccba['usgam'] > 100000) & (newccba['usgam'] <= 1000000), 'usgam' ] = 1 
newccba.loc[ (newccba['usgam'] > 1000000) & (newccba['usgam'] <= 5000000), 'usgam' ] = 2 
newccba.loc[ newccba['usgam'] > 5000000, 'usgam' ] = 3
newccba['usgam'].value_counts()
```

```python
newccba['inamt'].max()
newccba.loc[ newccba['inamt'] <= 100000, 'inamt' ] = 0 
newccba.loc[ (newccba['inamt'] > 100000) & (newccba['inamt'] <= 500000), 'inamt' ] = 1 
newccba.loc[ (newccba['inamt'] > 500000) & (newccba['inamt'] <= 1000000), 'inamt' ] = 2 
newccba.loc[ newccba['inamt'] > 1000000, 'inamt' ] = 3
newccba['inamt'].value_counts()
```
