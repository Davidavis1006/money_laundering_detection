```python
import pandas as pd
import numpy as np

data = pd.read_csv("public_train_x_cdtx0001_full_hashed.csv")
data = data.sort_values(by='cust_id')
```

```python
data
```

```python
#count the sum of amt
data_amt = data.groupby("cust_id")['amt'].mean()
```

```python
#count the duplicate of every different cust_id
count_dup=[]
x=0
for i in data.duplicated(subset=['cust_id']):
    if(i==False):
        if(x!=0):
            count_dup.append(x)
        x=1
    else:
        x=x+1
        
count_dup.append(x)
count_dup = pd.Series(count_dup,index=data_amt.index)
```

```python
#Use 0 to represent only trading in Taiwan, and 1 to represent internation trade
#Use 0 to represent only trading in NTD, and 1 to represent others
data['country'] = data['country'].map(lambda x:1 if x == 130 else 0)
data['cur_type'] = data['cur_type'].map(lambda x:1 if x == 47 else 0)
```

```python
data = data.groupby("cust_id")['country','cur_type'].sum()
data['amt']= data_amt

#add the column of trading times of every cust_id
data['times']=count_dup
#because using cust_id to be the index will have the problem that the index(cust_id) disapper when turning to the csv
#so we add one column that same as the index
data.insert(0,'cust_id',data.index)

data['country'] = data['country']/data['times']
data['cur_type'] = data['cur_type']/data['times']
```

```python
data.to_csv('cdtx.csv',index=False)
```
